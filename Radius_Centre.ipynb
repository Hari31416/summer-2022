```python
def r_and_c(image, threshold = 0.45, bound = 100):
    im = plt.imread(image)
    im = color.rgb2gray(im)
    im = im>threshold
    im = im.astype(int)
    
    x1 = im.max(axis=0).argmax()
    x2 = im.max(axis=0)[x1+1:].argmin() + x1 + 1
    x_c = int((x1+x2)/2)

    y1 = im.max(axis=1).argmax()
    y2 = im.max(axis=1)[y1+1:].argmin() + y1 + 1
    y_c = int((y1+y2)/2)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(im, cmap=plt.cm.gray)
    plt.scatter(x_c, y_c, c='g', s=20, marker='o', edgecolors='none', label='center')
    plt.vlines(x1, y_c-bound, y_c+bound, "w")
    plt.vlines(x2, y_c-bound, y_c+bound, "w")
    plt.hlines(y1, x_c-bound, x_c+bound, "w")
    plt.hlines(y2,x_c-bound, x_c+bound, "w")
    plt.legend()
    plt.axis('off')
    plt.show()

    r1 = np.abs(x_c-x1)
    r2 = np.abs(y_c-y1)
    r = int((r1+r2)/2)

    return x_c, y_c, r
```

```python
r_and_c("images/im1.png", 0.45, 100)
```

```python
r_and_c("images\im3.png")
```

### Edge Detection

```python
from skimage import filters
```

```python
image = "images\coin.jpg"
image = io.imread(image)
image = color.rgb2gray(image)
edge_roberts = filters.roberts(image)
edge_sobel = filters.sobel(image)

fig, axes = plt.subplots(ncols=2, sharex=True, sharey=True,
                         figsize=(8, 4))

axes[0].imshow(edge_roberts, cmap=plt.cm.gray)
axes[0].set_title('Roberts Edge Detection')

axes[1].imshow(edge_sobel, cmap=plt.cm.gray)
axes[1].set_title('Sobel Edge Detection')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
